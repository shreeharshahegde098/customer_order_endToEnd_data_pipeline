```python
%sql
CREATE OR REPLACE VIEW vw_customers AS
SELECT *
FROM delta.`/Volumes/workspace/default/shree_project/gold/customer_dim_delta`
WHERE is_current = true;
```

```python
%sql
CREATE OR REPLACE VIEW vw_orders AS
SELECT *
FROM delta.`/Volumes/workspace/default/shree_project/gold/order_fact`;
```

```python
%sql
CREATE OR REPLACE VIEW vw_products AS
SELECT *
FROM delta.`/Volumes/workspace/default/shree_project/gold/product_dim`;
```

```python
%sql
CREATE OR REPLACE VIEW vw_payments AS
SELECT *
FROM delta.`/Volumes/workspace/default/shree_project/gold/payment_dim`;
```

Revenue by city

```python
%sql
SELECT
    c.city,
    format_number(SUM(o.order_amount),0) AS total_revenue
FROM vw_orders o
JOIN vw_customers c
    ON o.customer_id = c.customer_id
GROUP BY c.city;
```

Orders by Revenue Category

```python
%sql
SELECT
    p.category,
    COUNT(o.order_id) AS total_orders,
    format_number(SUM(o.order_amount),0) AS revenue
FROM vw_orders o
JOIN vw_products p
    ON o.product_id = p.product_id
GROUP BY p.category;
```

Payments status Summary

```python
%sql
SELECT
    payment_status,
    COUNT(*) AS total_payments
FROM vw_payments
GROUP BY payment_status;
```

Daily Orders Trend

```python
%sql
SELECT
    order_date,
    COUNT(order_id) AS total_orders
FROM vw_orders
GROUP BY order_date
ORDER BY order_date;
```
